# Extract Australia Data from WHO COVID-19 Biweekly CSV.

```python
import pandas as pd

input_path = "../Data/WHO_COVID19_cleaned_biweekly.csv"   
output_path = "../Data/AUS_COVID19_biweekly.csv"

# Load the biweekly data
df = pd.read_csv(input_path)
print(f"Total rows: {df.shape[0]}")
print(f"Columns: {list(df.columns)}")
df.head()
```

```python
# Check Australia entries in the dataset
print(f"Country column unique matches: {df[df['Country'].str.contains('Australia', case=False, na=False)]['Country'].unique()}")
print(f"Country code: {df[df['Country'].str.contains('Australia', case=False, na=False)]['Country_code'].unique()}")
```

```python
aus_df = df[df['Country'] == 'Australia'].copy()
print(f"Australia rows: {aus_df.shape[0]}")
aus_df.head()
```

```python
aus_df.to_csv("../Data/AUS_COVID19_biweekly.csv", index=False)
print(f"Saved {aus_df.shape[0]} rows")
```
